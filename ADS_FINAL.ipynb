```python
binsedges = [0, 200, 500, 900, 2000, 3500,5000, 6900, 7600, 8000, 10000]
```

```python
flight_names= ['American Airlines', 'Delta Air Lines','United Airlines','Southwest Airlines','Air Canada','Air France','British Airways','Emirates','Lufthansa','Qatar Airways']
df["airlines"]=pd.cut(df['FlightNum'], bins=binsedges, labels= flight_names)
```

```python
df['airlines'].value_counts()
```

```python
#plot of delay for each carrier (stacked barchart)

#data prep for plot
carier_delay_perc_plot = pd.DataFrame() #empty df
carier_delay_perc_plot['airlines']=df['airlines'] #carrier variable with full names
carier_delay_perc_plot['ontime']=df['Total_Delay']<=0 #on time vs. delay
carier_delay_perc_plot = carier_delay_perc_plot.groupby(['airlines',
                                                           'ontime']).size().unstack(1) #shape and class data
carier_delay_perc_plot = carier_delay_perc_plot.div(carier_delay_perc_plot.sum(axis=1), axis=0) #percentage

#plot
f, ax = plt.subplots(figsize=(8, 6))
carier_delay_perc_plot.sort_values(by=False).plot(kind="bar", 
                                                  alpha=.9, stacked=True, ax=ax,color = palette) #stacked bar plot
#set labels
ax.set_title("Percentage of delay by airlines",fontsize = 30, fontname="Monospace", alpha=.8) #title
plt.xlabel('Carrier', fontsize = 16) #xlabel
ax.set_xticklabels(ax.get_xticklabels(),ha="right") #rotate xlabel
plt.ylabel("Percentage", fontsize = 16) #ylabel
ax.legend(loc='upper right', title='On time', fontsize = 13)

plt.show() #plot
```

```python
#create carrier total flights df
delay_carrier=df[['airlines']].groupby("airlines").count() #carrier column and group by count
delay_carrier['flights']=df['airlines'].value_counts() #total flights
delay_carrier = delay_carrier.reset_index(drop=False) #reset index

#carrier total delays df
delay_carrier2=(df[df['Total_Delay']>0]['airlines'].value_counts()) #delayed
delay_carrier2=delay_carrier2.to_frame(name=('Total_Delay')) #to df
delay_carrier2.index.name='airlines' #set index name
delay_carrier2=delay_carrier2.reset_index(drop=False) #reset index

#merge to carrier delay df
delay_merge = delay_carrier.merge(delay_carrier2[['airlines','Total_Delay']],
                                  how='left',left_on='airlines',right_on='airlines')
delay_merge['percentage']=(delay_merge['Total_Delay']/delay_merge['flights']) #% delay column
delay_merge #explore new df
```

```python
#multiple plots to gain better insight in delay by carrier.

#plot total flights by carrier (should show that Hawaiian has way less flights than big carriers)
f, ax = plt.subplots(figsize=(8, 6))
sns.barplot(data= delay_merge.sort_values(by='flights'),ax=ax, x=('airlines'), y=('flights'))
ax.set_title('Total flights by airlines',fontsize = 30, fontname="Monospace", alpha=.8)
ax.set_xticklabels(ax.get_xticklabels(), rotation=35, ha="right") #rotate xlabel
ax.set_xlabel('airlines',fontsize = 16)
ax.set_ylabel("Count (mio.)",fontsize = 16)
plt.tight_layout(rect=[0, 0, 1, 1])
plt.savefig('EDA_plot4.png', bbox_inches='tight') #save img
plt.show() #plot

```

```python
#plot total delays by carrier
f, ax = plt.subplots(figsize=(10, 8))
sns.barplot(data= delay_merge.sort_values(by='Total_Delay'),ax=ax, x=('airlines'), y=('Total_Delay'))
ax.set_title('Total delays by airlines')
ax.set_xticklabels(ax.get_xticklabels(), rotation=35, ha="right") #rotate xlabel
ax.set_xlabel('airlines')
ax.set_ylabel('No. of total delays')
plt.tight_layout() #tight layout
plt.show() #plot

```

```python
#ANOVA
import statsmodels.api as sm
from statsmodels.formula.api import ols


# Fit an ANOVA model to compare the mean delay times of different airlines
model = ols('Total_Delay ~ C(airlines)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print the ANOVA table
print(anova_table)
```
